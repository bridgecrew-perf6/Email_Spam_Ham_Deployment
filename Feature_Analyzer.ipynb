```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
tfidf_vect_fit = pickle.load(open("tfidf.pkl", 'rb'))
model_ = pickle.load(open("finalized_svm_model.sav", 'rb'))
```

```python
feature_names = tfidf_vect_fit.get_feature_names()
coefs_with_fns = sorted(zip(model_.coef_[0], feature_names)) 
df=pd.DataFrame(coefs_with_fns)
df.columns='coefficient','word'
df.sort_values(by='coefficient', ascending= False)[:50]
```
